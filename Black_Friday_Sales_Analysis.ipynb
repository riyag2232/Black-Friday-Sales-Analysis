# **Purchase Behavior Analysis:**
Analyze the distribution of the Purchase amount. Are there specific groups (e.g., age groups, gender, or city categories) that tend to spend more?



```python
age_distribution = df.groupby('Age')['Purchase'].sum()
age_distribution
```

```python
age_distribution.plot(kind='pie',autopct='%.0f%%')
plt.title('Purchase distribution by Age group')
plt.show()
```

```python
gender_distribution = df.groupby('Gender')['Purchase'].sum()
gender_distribution
```

```python
gender_distribution.plot(kind='pie',autopct='%.0f%%')
plt.title('Purchase distibution by Gender')
plt.show()
```

```python
city_distribution = df.groupby('City_Category')['Purchase'].sum()
city_distribution
```

```python
city_distribution.plot(kind='pie',autopct='%.0f%%')
plt.title('City Category wise Purchase distribution')
plt.show()
```

# **Customer Demographics:**

Analyze the distribution of customers based on Age, Gender, and City_Category. Which demographic group is most active during the Black Friday sales?



```python
age_dist = df.groupby('Age')['User_ID'].sum().reset_index().plot(kind='bar', x='Age', y='User_ID', color='pink', edgecolor='black')
plt.title('Customer distribution by Age')
plt.xlabel('Age')
plt.ylabel('count')
for i in age_dist.containers:
  age_dist.bar_label(i)
plt.xticks(rotation = 00)
plt.show()
```

```python
gender_dist = df.groupby('Gender')['User_ID'].sum().reset_index().plot(kind='bar', x='Gender', y='User_ID', color='brown', edgecolor='black', alpha=0.8)
plt.title('Customer distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
for i in gender_dist.containers:
  gender_dist.bar_label(i)
plt.xticks(rotation = 00)
plt.show()
```

```python
city_cate_dist = df.groupby('City_Category')['User_ID'].sum().reset_index().plot(kind='bar', x='City_Category', y='User_ID', color='purple', edgecolor='black', alpha=0.5)
plt.title('Customer distribution by city_category')
plt.xlabel('City_Category')
plt.ylabel('Count')
for i in city_cate_dist.containers:
  city_cate_dist.bar_label(i)
plt.xticks(rotation = 00)
plt.show()
```

# **Stay in Current City:**

Analyze how the duration of stay in the current city (Stay_In_Current_City_Years) affects purchase behavior. Do longer stays correlate with higher spending?


```python
current_city = df.groupby('Stay_In_Current_City_Years')['Purchase'].sum().reset_index()
current_city
```

```python
sns.barplot(x='Stay_In_Current_City_Years', y='Purchase', data=current_city, color='grey', edgecolor='black', alpha=0.5 )
plt.title('Purchase distribution by Stay_In_Current_City_Years')
plt.xlabel('Stay_In_Current_City_Years')
plt.ylabel('Purchase')
plt.xticks(rotation = 00)
plt.show()
```

By analyzing the duration of stay in the current city, found that customer who have stayed at least 1 year have more purchasing. This insight can help in customer retention and targeted marketing efforts.

# **Correlation Between Product Categories:**

Examine if there's any correlation between the three product categories (Product_Category_1, Product_Category_2, and Product_Category_3). Do customers who buy from one category tend to buy from another?


```python
product_cate = df[['Product_Category_1','Product_Category_2','Product_Category_3']]
product_cate.corr()
```

```python
sns.heatmap(product_cate.corr(),annot=True)
plt.title('Correlation heatmap of product distribution')
plt.xticks(rotation = 70)
plt.show()
```

A value close to 1 indicates a strong positive correlation there are customers who buy from one category tend to buy from another.

# **Gender-wise Product Preferences:**

Investigate if there are significant differences in product category preferences between male and female customers.

```python
prod_gen = df.groupby('Gender')['Product_Category_1'].value_counts()
prod_gen
```

```python
prod_gen.unstack().plot(kind='bar', label='Product_Category_1')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.title('product category preferred by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation = 00)
plt.show()
```
